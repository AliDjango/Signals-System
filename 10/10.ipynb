```
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    """
    Computes the step function for an array of values.

    Parameters:
        x (ndarray): The input values.

    Returns:
        ndarray: The step function for the input values.
    """
    return np.where(x >= 0, 1, 0)

# Generate the input values
x = np.linspace(-3, 3, 15)

# Compute the step function for 5 cycles
y = []
for i in range(5):
    y.append(step_function(x))

# Flatten the resulting list of arrays
y = np.concatenate(y)

# Compute the DTFT
N = len(y)
w = np.linspace(-np.pi, np.pi, N, endpoint=False)
Y = np.fft.fftshift(np.fft.fft(y))

# Plot the results
plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
plt.stem(y)
plt.xlabel('n')
plt.ylabel('y[n]')
plt.title('Step Function ')

plt.subplot(2, 1, 2)
plt.plot(w, np.abs(Y))
plt.xlabel('Frequency (radians/sample)')
plt.ylabel('|Y(w)|')
plt.title('DTFT of Step Function ')
plt.tight_layout()
plt.show()
#Ali Eshraghi Amirhossein Karami Agha Reza Sobhani 10.9
```
