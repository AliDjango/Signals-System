**Example - Fourier transform of causal exponential signal**

The Fourier transform $X(j \omega) = \mathcal{F} \{ x(t) \}$ of the causal exponential signal

\begin{equation}
x(t) = \epsilon(t) \cdot e^{- \alpha t}
\end{equation}

with $\alpha \in \mathbb{R}^+$ is derived from the Laplace transform of the [causal complex exponential signal](#Transformation-of-the-Causal-Complex-Exponential-Signal). Using the substitution $s_0 = \alpha$ yields

\begin{equation}
X(s) = \frac{1}{s + \alpha}
\end{equation}

with the ROC $\Re \{ s \} > - \alpha$. Hence for $0 < \alpha < \infty$, the ROC includes the imaginary axis. In this case the signal comprises an exponential signal whose level decays over time for $t>0$. The Fourier transform can be derived from the Laplace transform as

\begin{equation}
X(j \omega) = \frac{1}{j \omega + \alpha}
\end{equation}

```

import numpy as np
from scipy import fft

# Define the causal exponential signal
def f(t, a):
    return np.where(t < 0, 0, np.exp(-a*t))

# Set the value of "a" and the range of time values
a = 10
t = np.linspace(-5, 1, 500)

# Calculate the Fourier transform using the FFT algorithm
dt = t[1] - t[0]
Fw = fft.fft(f(t, a)) * dt
w = fft.fftfreq(len(t), dt) * 2 * np.pi

# Plot the causal exponential signal and its Fourier transform
import matplotlib.pyplot as plt
plt.subplot(211)
plt.plot(t, f(t, a))
plt.title('Causal Exponential Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.subplot(212)
plt.plot(w, np.abs(Fw))
plt.title('Fourier Transform of Causal Exponential Signal')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from scipy import integrate

pi = np.pi

t = np.linspace(-10,11,500)

T_0 = 4
omega0 = 2*pi/T_0

y_axis = signal.square(omega0 * t)
y_axis = (y_axis/2) + 0.5

fft_result = np.fft.fft(y_axis)

a_k = fft_result/T_0

omega = t * omega0

dt = t[1] - t[0]
freqs = np.fft.fftfreq(len(y_axis), dt)

plt.subplot(2, 1, 1)
plt.plot(t-1, y_axis)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('square Wave')
plt.grid()

plt.subplot(2, 1, 2)
plt.stem(freqs, np.abs(a_k))
plt.xlabel('Frequency')
plt.xlim(-5,5)
plt.ylabel('Amplitude')
plt.title('a_k = X(ω)/T_0')
plt.grid()

plt.tight_layout()
```

```
import numpy as np
import matplotlib.pyplot as plt

# Define the symmetric square wave
T = 2  # set the period of the square wave
t = np.linspace(-T/2, T/2, 1000)
x = np.zeros(len(t))
for i in range(len(t)):
    if np.abs(t[i]) < T/4:
        x[i] = 1
    elif np.abs(t[i]) > T/4:
        x[i] = -1

# Compute the Fourier series coefficients
n = np.arange(-10, 11)
c = np.zeros(len(n), dtype=complex)
for i in range(len(n)):
    c[i] = (1/T) * np.trapz(x * np.exp(-1j * 2 * np.pi * n[i] * t / T), t)

# Compute the Fourier transform
X = np.fft.fft(x)
freq = np.fft.fftfreq(len(t), t[1]-t[0])

# Plot the results
plt.subplot(311)
plt.plot(t, x)
plt.title('Symmetric square wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(312)
plt.stem(n, c.real)
plt.title('Fourier series coefficients')
plt.xlabel('Harmonic')
plt.ylabel('Amplitude')

plt.subplot(313)
plt.plot(freq, np.abs(X))
plt.title('Fourier transform')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')

plt.show()
#aydin raisian iman rafiee alireza rouzbahani
```
