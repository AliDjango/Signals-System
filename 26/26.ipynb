```
import numpy as np
import matplotlib.pyplot as plt
def arctan2(w, T):
    return 2 * np.arctan(w * T / 2)
w = 1
T = np.linspace(0, 10, 1000)
y = arctan2(w, T)

plt.plot(T, y)
plt.xlabel('Time')
plt.ylabel('Angle')
plt.title('2arctan(wT/2)')
plt.show()
```
