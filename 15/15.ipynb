```
#15.6
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal


num_samples = 100
period = 8


samples = np.arange(num_samples)

impulse_train = signal.unit_impulse(num_samples, samples % period == 0)


plt.stem(samples, impulse_train)
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# Define the function f(x) for discrete values of x with larger spacing between dots and ones from 0 to 4
x1 = np.arange(-10, -6, 0.5)
x2 = np.arange(0, 4, 0.5)
x = np.concatenate((x1, x2))
y = np.zeros(len(x))
y[(x >= 0) & (x < 2.5)] = 1
y[(x >= 2.5) & (x < 4)] = 1

# Plot the function using stems
plt.figure(figsize=(8, 5))
plt.stem(x, y, linefmt='C0-', markerfmt='C0o', basefmt='k', use_line_collection=True)

# Set the x and y axis labels and the plot title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = {0, x < 0; 1, 0 <= x < 2.5; 1, 2.5 <= x < 4; 0, x >= 4}')

# Show the plot
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-1, 1, 1000)

triangle_pulse = np.zeros_like(t)
triangle_pulse[np.abs(t) <= 0.5] = 1 - 2 * np.abs(t[np.abs(t) <= 0.5])

plt.plot(t, triangle_pulse)
plt.xlabel('')
plt.ylabel('X()')
plt.grid(True)
plt.show()
```
