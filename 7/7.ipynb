```
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
#define a_k coefficients function
def a_k(k):
  if k==0: result = 1/2
  else: result = ( np.sin(k*pi/2) ) / ( k*pi )
  return result
#define pi in easy way
pi = np.pi
#set min and max value of Time-axis
x_min_val = -10
x_max_val = 10

x_axis = np.linspace(x_min_val, x_max_val, num=500)     #define an array for our points in x-axis
y = []    #define empty list of y-axis

T_0 = 2*pi           #T_0 is period amount
omega0 = 2*pi / T_0

a_k_list = []              #this list defined for values of π*a_k special plot
k_axis = np.arange(0,11,1)   #define k-axis points
for k in k_axis:
    a_k_list.append(a_k(k))    #fill k-axis list
#mirror points value vertically
k_axis = np.arange(-10,11,1)
pixa_k = pi * np.array(a_k_list)
pixa_k = np.append(np.flipud(pixa_k[1:]),pixa_k)

for t in x_axis:          #this loop fill y-axis list with each point in x-axis defined as parameter t
  result = a_k(0)
  for k in range(1, 1000, 1):    #this loop is sigma in fourier-series
    result += 2 * a_k(k) * np.cos(k*omega0*t )
  y.append(result)        #put y-axis in list

y_axis = np.array(y)       #convert y-axis list to an array

#plot figure 1 as Symmetric Periodic Square Wave
plt.figure(1)
plt.plot(x_axis, y_axis)
plt.title('Symmetric Periodic Square Wave\n(Fourier-Series)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.ylim(-0.5,2)
plt.grid()
#plot figure 2
plt.figure(2)
plt.plot(k_axis, pixa_k)
plt.xlabel('k')
plt.ylabel('πa_k')
plt.xticks(np.arange(-10,10,1))
plt.ylim(-0.5,2)
plt.grid()

```

```
import numpy as np
import matplotlib.pyplot as plt

# Define the Fourier eigenfunction
T0 = 10  # set the period of the signal
f = 1/T0  # compute the frequency
t = np.linspace(-T0/2, T0/2, 1000)
x = np.exp(1j * 2 * np.pi * f * t)

# Compute the Fourier transform
X = np.fft.fft(x)
freq = np.fft.fftfreq(len(t), t[1]-t[0])

# Compute the complex exponential representation
X_complex = np.zeros(len(X), dtype=complex)
for i in range(len(X)):
    X_complex[i] = X[i] * np.exp(2j * np.pi * freq[i] * t[0])

# Plot the results
plt.subplot(211)
plt.plot(t, x.real, label='Real part')
plt.plot(t, x.imag, label='Imaginary part')
plt.legend()

plt.subplot(212)
plt.plot(freq, X_complex.real, label='Real part')
plt.plot(freq, X_complex.imag, label='Imaginary part')
plt.legend()

plt.show()
#aydin raisian iman rafiee alireza rouzbahani
```

```

import numpy as np
import matplotlib.pyplot as plt

def f(t, T0):
    if T0 == np.inf:
        return np.ones_like(t)
    else:
        return np.abs(t)<T0/2

T0 = 2*np.pi
t = np.linspace(-5*T0, 5*T0, 1000)
plt.plot(t, f(t, T0))
plt.xlabel('t')
plt.ylabel('f(t)')

plt.grid()
plt.show()
```
