```
#19.1
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal


num_samples = 100
period = 5


samples = np.arange(num_samples)


impulse_train = np.zeros(num_samples)
impulse_train[samples % period == 0] = 1
impulse_train[samples % (2 * period) == 0] = 2
impulse_train[samples % (3 * period) == 0] = 3
impulse_train[samples % (4 * period) == 0] = 4
impulse_train[samples % (5 * period) == 0] = 5




plt.stem(samples, impulse_train)
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np


def f(x):
    return -np.abs(x) + 0.5
x1 = np.linspace(-4, -2, 1000)
x2 = np.linspace(-2, 0, 1000)
x3 = np.linspace(0, 2, 1000)
x4 = np.linspace(2, 4, 1000)
fig, ax = plt.subplots()
ax.plot(x1, f(x1), label='[-4, -2]')
ax.plot(x2, f(x2), label='[-2, 0]')
ax.plot(x3, f(x3), label='[0, 2]')
ax.plot(x4, f(x4), label='[2, 4]')
ax.plot(x1+8, f(x1), label='[-4, -2]')
ax.plot(x2+8, f(x2), label='[-2, 0]')
ax.plot(x3+8, f(x3), label='[0, 2]')
ax.plot(x4+8, f(x4), label='[2, 4]')
ax.plot(x1-8, f(x1), label='[-4, -2]')
ax.plot(x2-8, f(x2), label='[-2, 0]')
ax.plot(x3-8, f(x3), label='[0, 2]')
ax.plot(x4-8, f(x4), label='[2, 4]')

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Plot of f(x) = -|x| + 1/2')
ax.legend()

plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# Define the function f(x) for discrete values of x with larger spacing between dots and ones from 0 to 5
x = np.arange(-10, 10, 0.5)
y = np.zeros(len(x))
y[(x >= 0) & (x < 2.5)] = 1
y[(x >= 2.5) & (x < 5)] = 1

# Plot the function using stems
plt.figure(figsize=(8, 5))
plt.stem(x, y, linefmt='C0-', markerfmt='C0o', basefmt='k', use_line_collection=True)

# Set the x and y axis labels and the plot title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = {0, x < 0; 1, 0 <= x < 2.5; 1, 2.5 <= x < 5; 0, x >= 5}')

# Show the plot
plt.show()
#Agha Reza Sobhani ,  Ali Eshraghi . Amirhossein karami 19.15
```

```
import matplotlib.pyplot as plt
import numpy as np

# Define the function f(x) for discrete values of x with dots and larger spacing between dots
x = np.arange(-10, 10, 0.5)
y = np.zeros(len(x))
y[x < 0] = 0
y[(x >= 0) & (x < 2*np.pi)] = np.abs(np.sin(x[(x >= 0) & (x < 2*np.pi)])) + 0.5
y[x >= 2*np.pi] = 0

# Plot the function using stems with dots and larger spacing between dots
plt.figure(figsize=(8, 5))
plt.stem(x, y, linefmt='k', markerfmt='ko', basefmt='k', use_line_collection=True)

# Set the x and y axis labels and the plot title
plt.xlabel('x')
plt.ylabel('f(x)')
# Show the plot
plt.show()
#Agha Reza Sobhani ,  Ali Eshraghi . Amirhossein karami 19.15
```

```
import matplotlib.pyplot as plt
import numpy as np

# Define the function f(x) for discrete values of x with dots and larger spacing between dots
x = np.arange(-22, 22, 0.5)
y = np.zeros(len(x))
y[(x >= -5) & (x < 5)] = np.abs(np.sin(x[(x >= -5) & (x < 5)])) + 0.5
y[(x >= -10) & (x < 0)] = 0

# Plot the function using stems with dots and larger spacing between dots
plt.figure(figsize=(10, 5))
plt.stem(x, y, linefmt='k', markerfmt='ko', basefmt='k', use_line_collection=True)

# Add the dots from 0 to 5 in the range -20 to -15
plt.stem(np.arange(-20, -15, 0.5), np.abs(np.sin(np.arange(-20, -15, 0.5))) + 0.5, linefmt='k', markerfmt='ko', basefmt='k', use_line_collection=True)

# Add the dots from 0 to 5 in the range 15 to 20
plt.stem(np.arange(15, 20, 0.5), np.abs(np.sin(np.arange(15, 20, 0.5))) + 0.5, linefmt='k', markerfmt='ko', basefmt='k', use_line_collection=True)

# Set the x and y axis labels and the plot title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = {0, -10 <= x < 0 or -5 <= x < 5; |sin(x)| + 0.5, 0 <= x < 2pi; 0, otherwise}')

# Show the plot
plt.show()
#Agha Reza Sobhani ,  Ali Eshraghi . Amirhossein karami 19.15
```
