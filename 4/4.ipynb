\begin{equation}
x[n] = \sum_{\kappa = -\infty}^{\infty} x[\kappa] \cdot \delta[n - k]\\y[n] = \sum_{\kappa = -\infty}^{\infty} x[\kappa] \cdot h_k[n]\\\delta[n-k]=h_k[n] \\if\ time\ invariant:\\h_k[n]=h_0[n-k]\\LTI:y[n] = \sum_{\kappa = -\infty}^{\infty} x[\kappa] \cdot h[n-k]
\end{equation}

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def heaviside(k):
    return np.where(k >= 0, 1.0, 0.0)


def rect(k, N):
    return np.where((0 <= k) & (k < N), 1.0, 0.0)


def x(k):
    return 4/5 * rect(k, 6)


def h(k):
    return heaviside(k) * np.exp(- k/2)


def plot_signals(k, x, h, xlabel, hlabel, klabel):
    plt.figure(figsize=(8, 4))
    plt.stem(k, x, linefmt='C0-', markerfmt='C0o', label=xlabel)
    plt.stem(k, h, linefmt='C1-', markerfmt='C1o', label=hlabel)
    plt.xlabel(klabel)
    plt.legend()
    plt.ylim([0, 1.2])
    plt.grid(True)
```

```
k = np.arange(-20, 21)

plot_signals(k, x(k), h(k), r'$x[k]$', r'$h[k]$', r'$k$')
```
