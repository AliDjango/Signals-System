```
import numpy as np
import matplotlib.pyplot as plt

# Define the system parameters
a = 0.5
n = np.arange(0, 20)
h = a**n  # Impulse response

# Compute the frequency response
w = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
H = 1 / (1 - a * np.exp(-1j * w))

# Plot the impulse response
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))
ax1.stem(n, h)
ax1.set_xlabel('n')
ax1.set_ylabel('Amplitude')
ax1.set_title(f'Impulse Response, a = {a}')

# Plot the frequency response
f = w / (2 * np.pi)
ax2.plot(f, np.abs(H))
ax2.set_xlabel('Frequency (Hz)')
ax2.set_ylabel('Magnitude')
ax2.set_ylim([0, 10])
ax2.set_title(f'Frequency Response, a = {a}')
ax2.set_xlim([-1, 1])

plt.tight_layout()
plt.show()
#Ali Eshraghi Amirhossein Karami Agha Reza Sobhani 11.10
```
