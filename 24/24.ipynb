```
#24.7
from scipy import signal
import matplotlib.pyplot as plt

# define the order of the filter and the critical frequency
order = 4
Wn = 0.2

# create the Butterworth filter
b, a = signal.butter(order, Wn, 'low')

# create the pole-zero plot
z, p, k = signal.tf2zpk(b, a)
plt.figure()
plt.plot(np.real(z), np.imag(z), 'xb')
plt.plot(np.real(p), np.imag(p), 'or')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.grid()
plt.title('Pole-Zero Plot of Butterworth Filter')

# map the filter to the desired discrete-time filter
dt_b, dt_a = signal.bilinear(b, a, fs=1)

# create the pole-zero plot of the discrete-time filter
dt_z, dt_p, dt_k = signal.tf2zpk(dt_b, dt_a)
plt.figure()
plt.plot(np.real(dt_z), np.imag(dt_z), 'xb')
plt.plot(np.real(dt_p), np.imag(dt_p), 'or')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.grid()
plt.title('Pole-Zero Plot of Discrete-Time Filter')

# show the plots
plt.show()
```
