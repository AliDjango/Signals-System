```
import numpy as np
import matplotlib.pyplot as plt

# Define the Fourier eigenfunction
T0 = 10  # set the period of the signal
f = 1/T0  # compute the frequency
t = np.linspace(-T0/2, T0/2, 1000)
x = np.exp(1j * 2 * np.pi * f * t)

# Compute the Fourier transform
X = np.fft.fft(x)
freq = np.fft.fftfreq(len(t), t[1]-t[0])

# Compute the complex exponential representation
X_complex = np.zeros(len(X), dtype=complex)
for i in range(len(X)):
    X_complex[i] = X[i] * np.exp(2j * np.pi * freq[i] * t[0])

# Plot the results
plt.subplot(211)
plt.plot(t, x.real, label='Real part')
plt.plot(t, x.imag, label='Imaginary part')
plt.legend()

plt.subplot(212)
plt.plot(freq, X_complex.real, label='Real part')
plt.plot(freq, X_complex.imag, label='Imaginary part')
plt.legend()

plt.show()
#aydin raisian iman rafiee alireza rouzbahani
```

```
import numpy as np
import matplotlib.pyplot as plt

# Define the symmetric square wave
T = 2  # set the period of the square wave
t = np.linspace(-T/2, T/2, 1000)
x = np.zeros(len(t))
for i in range(len(t)):
    if np.abs(t[i]) < T/4:
        x[i] = 1
    elif np.abs(t[i]) > T/4:
        x[i] = -1

# Compute the Fourier series coefficients
n = np.arange(-10, 11)
c = np.zeros(len(n), dtype=complex)
for i in range(len(n)):
    c[i] = (1/T) * np.trapz(x * np.exp(-1j * 2 * np.pi * n[i] * t / T), t)

# Compute the Fourier transform
X = np.fft.fft(x)
freq = np.fft.fftfreq(len(t), t[1]-t[0])

# Plot the results
plt.subplot(311)
plt.plot(t, x)
plt.title('Symmetric square wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(312)
plt.stem(n, c.real)
plt.title('Fourier series coefficients')
plt.xlabel('Harmonic')
plt.ylabel('Amplitude')

plt.subplot(313)
plt.plot(freq, np.abs(X))
plt.title('Fourier transform')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')

plt.show()
#aydin raisian iman rafiee alireza rouzbahani
```

```
import numpy as np
import matplotlib.pyplot as plt

# Define the discrete-time periodic signal
N = 50  # set the period of the signal
n = np.arange(N)
omega = 2*np.pi/N  # set the fundamental frequency
x = 1 + np.sin(omega*n) + 3*np.cos(omega*n) + np.cos(2*omega*n + np.pi/2)

# Compute the Fourier series coefficients
k = np.arange(-N/2, N/2)
c = np.zeros(len(k), dtype=complex)
for i in range(len(k)):
    c[i] = np.sum(x * np.exp(-1j * 2 * np.pi * k[i] * n / N)) / N

# Plot the results
plt.subplot(311)
plt.stem(n, x)
plt.title('Discrete-time periodic signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(312)
plt.stem(k, c.real)
plt.title('Real part of Fourier series coefficients')
plt.xlabel('Harmonic')
plt.ylabel('Amplitude')

plt.subplot(313)
plt.stem(k, c.imag)
plt.title('Imaginary part of Fourier series coefficients')
plt.xlabel('Harmonic')
plt.ylabel('Amplitude')

plt.show()
#aydin raisian iman rafiee alireza rouzbahani
```

```
import numpy as np
import matplotlib.pyplot as plt
# matplotlib inline
def exponential_signal(k,sigma,omega):
  return np.exp((sigma + 1j* omega)*k)
def plot_signal(k,x):
     plt.figure(figsize=(10,4))
     plt.subplot(121)
     plt.stem(k,np.real(x))
     plt.xlabel('$k$')
     plt.ylabel(r'$\Re\{x[k]\}$')

     plt.subplot(122)
     plt.stem(k,np.imag(x))
     plt.xlabel('$k$')
     plt.ylabel(r'$\Im\{x[k]\}$')
     plt.tight_layout()
```
