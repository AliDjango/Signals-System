```
import matplotlib.pyplot as plt
import numpy as np


def f(x):
    return -np.abs(x) + 0.5
x1 = np.linspace(-4, -2, 1000)
x2 = np.linspace(-2, 0, 1000)
x3 = np.linspace(0, 2, 1000)
x4 = np.linspace(2, 4, 1000)
fig, ax = plt.subplots()
ax.plot(x1, f(x1), label='[-4, -2]')
ax.plot(x2, f(x2), label='[-2, 0]')
ax.plot(x3, f(x3), label='[0, 2]')
ax.plot(x4, f(x4), label='[2, 4]')
ax.plot(x1+8, f(x1), label='[-4, -2]')
ax.plot(x2+8, f(x2), label='[-2, 0]')
ax.plot(x3+8, f(x3), label='[0, 2]')
ax.plot(x4+8, f(x4), label='[2, 4]')
ax.plot(x1-8, f(x1), label='[-4, -2]')
ax.plot(x2-8, f(x2), label='[-2, 0]')
ax.plot(x3-8, f(x3), label='[0, 2]')
ax.plot(x4-8, f(x4), label='[2, 4]')

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Plot of f(x) = -|x| + 1/2')
ax.legend()

plt.show()
#Agha Reza Sobhani ,  Ali Eshraghi . Amirhossein karami 17.1
```

```
import matplotlib.pyplot as plt
import numpy as np

# Define the function f(x)
def f(x):
    if x < -1:
        return 0
    elif x >= -2 and x < 2:
        return 1
    else:
        return 0

# Create an array of x values
x = np.linspace(-5, 5, 2000)

# Calculate the corresponding y values for each x value
y = [f(i) for i in x]

# Plot the function
plt.plot(x, y)

# Set the x and y axis labels and the plot title
plt.xlabel('x')
plt.ylabel('f(x)')

# Show the plot
plt.show()
#Agha Reza Sobhani ,  Ali Eshraghi . Amirhossein karami 17.1
```

```
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

N = 5

x = np.convolve(np.ones(N), np.ones(N), mode='full')
plt.stem(x)
plt.xlabel('$k$')
plt.ylabel('$x[k]$')

W = sym.symbols('Omega')
X = sym.exp(-sym.I*W * (N-1)) * sym.sin(N*W/2) / sym.sin(W/2)
sym.plot(sym.Abs(X), (W, -2, 2), xlabel='$\Omega$',
         ylabel='$|X(e^{j \Omega})|$')
```
