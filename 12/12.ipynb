```
import numpy as np
import matplotlib.pyplot as plt

# Define filter parameters
fc = 100  # Cutoff frequency
fs = 1000  # Sampling frequency

# Create frequency vector
num_points = 10000
freq = np.linspace(0, fs/2, num=num_points)

# Generate ideal low-pass filter frequency response
H = np.where((freq >= 200) & (freq <= 300), 1, 0)

# Plot frequency response
plt.plot(freq, H)
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.title('Ideal low-pass filter frequency response')
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# Define filter parameters
fc = 100  # Cutoff frequency
fs = 1000  # Sampling frequency

# Create frequency vector
num_points = 10000
freq = np.linspace(0, fs/2, num=num_points)

# Generate ideal high-pass filter frequency response
H = np.where((freq < 100) | ((freq > 200) & (freq < 300)) | (freq > 400), 1, 0)

# Plot frequency response
plt.plot(freq, H)
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.title('Ideal high-pass filter frequency response')
plt.show()
```

```
import scipy.signal as sig
import matplotlib.pyplot as plt
import numpy as np

# compute transfer function
a = (1, -2, 1/2)  # convention: a = (a0, a1, a2,...)
b = (1)  # convention: b = (b0, b1, b2,...)
W, H = sig.freqz(b, a, whole=True, worN=np.linspace(-np.pi, np.pi, 2**10))

# plot transfer function
plt.figure()
plt.plot(W, abs(H))
plt.ylabel(r'$|H(e^{j \Omega})|$')
plt.xlabel(r'$\Omega$')
plt.ylim([0, 3])
plt.xlim([-np.pi, np.pi])
plt.grid()
```
