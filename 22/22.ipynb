```
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Define the exponential sequence
a = 0.8
n = np.arange(-10, 1)
x = a**(-n)

# Compute the poles and zeros of the transfer function
zeros = [0]
poles = [a]
gain = 1

# Plot the pole-zero plot
zplane = plt.figure(figsize=(6,6))
ax = zplane.add_subplot(111)
ax.scatter(np.real(zeros), np.imag(zeros), marker='o', s=100, facecolors='none', edgecolors='b')
ax.scatter(np.real(poles), np.imag(poles), marker='x', s=100, color='r')
circ = plt.Circle((0,0), radius=1, linestyle='--', fill=False, color='k')
ax.add_patch(circ)
plt.axis('scaled')
plt.xlim([-1.2, 1.2])
plt.ylim([-1.2, 1.2])
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Pole-Zero Plot')
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# Define the function f(x)
def f(x):
    if x < -1:
        return np.exp(x)
    elif x >= -1 and x < 1:
        return x**2 + 2
    else:
        return np.exp(-x)

# Create an array of x values
x = np.linspace(-5, 5, 1000)

# Calculate the corresponding y values for each x value
y = [f(i) for i in x]

# Plot the function
plt.plot(x, y)

# Set the x and y axis labels and the plot title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = {e^x, x < -1; x^2+2, -1 <= x < 1; e^-x, x >= 1}')

# Show the plot
plt.show()
#Agha Reza Sobhani ,  Ali Eshraghi . Amirhossein karami 22.4
```
